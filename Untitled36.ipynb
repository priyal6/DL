```python
import torch
import math
import matplotlib.pyplot as plt

class SinusoidalPositionalEncoding(torch.nn.Module):
    def __init__(self, d_model, max_len=5000):
        super().__init__()

        # Create a matrix of [max_len, d_model] representing each position's encoding
        pe = torch.zeros(max_len, d_model)
        position = torch.arange(0, max_len).unsqueeze(1).float()  # Shape: [max_len, 1]
        div_term = torch.exp(torch.arange(0, d_model, 2).float() * (-math.log(10000.0) / d_model))

        pe[:, 0::2] = torch.sin(position * div_term)  # Even indices
        pe[:, 1::2] = torch.cos(position * div_term)  # Odd indices

        pe = pe.unsqueeze(0)  # Shape: [1, max_len, d_model]
        self.register_buffer('pe', pe)  # Will not be updated during training

    def forward(self, x):
        """
        x: Tensor of shape [batch_size, seq_len, d_model]
        """
        seq_len = x.size(1)
        return x + self.pe[:, :seq_len, :]
```

```python
def plot_positional_encodings(pe, num_dimensions=6):
    plt.figure(figsize=(12, 6))
    position = torch.arange(0, pe.size(1))

    for i in range(num_dimensions):
        plt.plot(position, pe[0, :, i].numpy(), label=f'Dim {i}')

    plt.legend()
    plt.title("Sinusoidal Positional Encoding")
    plt.xlabel("Position")
    plt.ylabel("Encoding Value")
    plt.grid(True)
    plt.show()

# Example usage
d_model = 32
max_len = 100
pe_module = SinusoidalPositionalEncoding(d_model, max_len)
plot_positional_encodings(pe_module.pe, num_dimensions=6)
```
